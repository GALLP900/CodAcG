```python
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
from itertools import product
```

```python
#define the AND gate
data = [[0,0], [0,1], [1,0], [1,1]]
labels = [0,0,0,1]
#fit the perceptron model and print score
classifier = Perceptron(max_iter = 40, random_state=22)
classifier.fit(data, labels)
print(classifier.score(data, labels))
print(classifier.decision_function([[0, 0], [1, 1], [0.5, 0.5]]))
#create a list of predictions
x_values = np.linspace(0,1,100)
y_values = np.linspace(0,1,100)
point_grid = list(product(x_values,y_values))
distances = classifier.decision_function(point_grid)
abs_distances = abs(distances)
#reshape the predictions into a two dimensional list of points
distances_matrix = np.reshape(abs_distances,(100,100))
#plot the distances matrix
heatmap = plt.pcolormesh(x_values,y_values,distances_matrix, cmap='viridis', animated=True, alpha=0.8)
plt.colorbar(heatmap)
plt.scatter(x=[point[0] for point in data],y=[point[1] for point in data],c=labels)
plt.show()
plt.close()
```
