```python
import numpy as np
import sympy as sp
from math import sin, cos, log, pi
from sympy import pi, sec, simplify
import matplotlib.pyplot as plt

def limit_derivative(f, x, h):
  """
  f: function to be differentiated 
  x: the point at which to differentiate f 
  h: distance between the points to be evaluated
  """
  #compute the derivative at x with limit definition
  if h == 0:
    return "Error: h value should not be zero"
  else:
    return (f(x + h) - f(x)) / h

# f1(x) = sin(x)
def f1(x):
    return np.sin(x)

# f2(x) = x^4
def f2(x):
    return pow(x, 4)

# f3(x) = x^2*log(x)
def f3(x):
    return pow(x, 2) * np.log(x)

def f4(x):
    return np.tan(x)
# Graph the true derivative
#f1
x_vals = np.linspace(1, 10, 1000)
#f2
#y_vals = [cos(val) for val in x_vals]
#f2
#y_vals = [4*pow(val,3) for val in x_vals]
#f4
#y_vals = [pow(val, 2) * log(val) for val in x_vals]
#f4
y_vals = [simplify(sec(val)) for val in x_vals]
plt.figure(1)
plt.plot(x_vals, y_vals, label="True Derivative", linewidth=2)

# plot different approximated derivatives of f using numpy gradient
def plot_approx_deriv_numpy(f):
  x_vals = np.linspace(1, 10, 1000)
  h_vals = [2, 1, .5, .08, 0.04]
  for h in h_vals:
      y_val = np.gradient(f(x_vals),h)
      plt.plot(x_vals, y_val, linestyle='--', label="h=" + str(h) )
  plt.legend()
  plt.title("Convergence to Derivative by varying h")
  plt.ylim(-50, 50)
  plt.xlim(1, 2)
  plt.show()

# Plot here
plot_approx_deriv_numpy(f4)
```
